```python
'''
This is my code for the Lorenz Attractor,
copied over from my learning portfolio.
'''

# constants
s = 10
b = 8 / 3
p = 28

# time
time = 0
max_time = 50
dt = 0.0001 # change this when needed

# variables
x = 0
y = 1
z = 0
data = []

def xt(x, y):
    return s * (y - x) * dt

def yt(x, y, z):
    return (x * (p - z) - y) * dt

def zt(x, y, z):
    return ((x * y) - (b * z)) * dt

while time < max_time:
    data.append([x, y, z])
    x = x + xt(x, y)
    y = y + yt(x, y, z)
    z = z + zt(x, y, z)
    time += dt

data = np.array(data)
data = {
    'x': data[:, 0],
    'y': data[:, 1],
    'z': data[:, 2],
}

ax = plt.axes(projection='3d')
ax.plot3D(data['x'], data['y'], data['z'])
plt.show()
```
